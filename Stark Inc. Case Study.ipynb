```python
# Calculate the frequency of each unique impression value
impression_counts = All_Raw_User_IDs_Summarized_ByAllValues['impressions'].value_counts()

# Create a new DataFrame to summarize the values
Install_Raw_User_IDs_Summarized_Impression_Frequency = pd.DataFrame({'Impressions': impression_counts.index, 'Frequency': impression_counts.values})

# Sort the DataFrame by the 'Impressions' column
Install_Raw_User_IDs_Summarized_Impression_Frequency = Install_Raw_User_IDs_Summarized_Impression_Frequency.sort_values('Impressions')

# Display the summary DataFrame
print(Install_Raw_User_IDs_Summarized_Impression_Frequency)
```

```python
# Order the DataFrame by a specific column in descending order
Install_Raw_User_IDs_Summarized_Impression_Frequency = Install_Raw_User_IDs_Summarized_Impression_Frequency.sort_values('Frequency', ascending=False)

# Display the ordered DataFrame
print(Install_Raw_User_IDs_Summarized_Impression_Frequency)
```

```python
# Calculate the column total
column_total = Install_Raw_User_IDs_Summarized_Impression_Frequency['Frequency'].sum()

# Calculate the percentage for each value in the column
Install_Raw_User_IDs_Summarized_Impression_Frequency['Percentage'] = Install_Raw_User_IDs_Summarized_Impression_Frequency['Frequency'].apply(lambda x: (x / column_total) * 100)
```

```python
Install_Raw_User_IDs_Summarized_Impression_Frequency
```

```python
# Calculate the weighted average
weighted_avg = np.average(Install_Raw_User_IDs_Summarized_Impression_Frequency['Impressions'], weights=Install_Raw_User_IDs_Summarized_Impression_Frequency['Frequency'])

# Print the weighted average
print(weighted_avg)
```

```python
# Exclude the first two indices from values and weights
values_excluded = Install_Raw_User_IDs_Summarized_Impression_Frequency['Impressions'][2:]
weights_excluded = Install_Raw_User_IDs_Summarized_Impression_Frequency['Frequency'][2:]

# Calculate the total weight
total_weight = np.sum(weights_excluded)

# Calculate the weighted average
weighted_average = np.average(values_excluded, weights=weights_excluded)

print("1st 80% Excluded Weighted Average:", weighted_average)
```
